```python
#Images
import os
import numpy as np 
import matplotlib.pyplot as plt 
import matplotlib.image as mpimg 
%matplotlib inline
```

```python
#importing and viewing all images from folder
img_filenames = sorted(os.listdir('C:/Users/hruss/OneDrive/Documents/GMU/CSI-703/CSI703/The-Data-Visualization-Workshop/Datasets/images')) 
imgs = [mpimg.imread(os.path.join('C:/Users/hruss/OneDrive/Documents/GMU/CSI-703/CSI703/The-Data-Visualization-Workshop/Datasets/images', img_filename)) for img_filename in img_filenames]
fig, axes = plt.subplots(1, 4) 
fig.figsize = (6, 6) 
fig.dpi = 150 
axes = axes.ravel() 
# Specify labels 
labels = ['coast', 'beach', 'building', 'city at night'] 
# Plot images 
for i in range(len(imgs)): 
    axes[i].imshow(imgs[i]) 
    axes[i].set_xticks([]) 
    axes[i].set_yticks([]) 
    axes[i].set_xlabel(labels[i], color='white')
```
