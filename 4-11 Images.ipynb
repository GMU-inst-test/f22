```python
#Import libraries
import os 
import numpy as np 
import matplotlib.pyplot as plt 
import matplotlib.image as mpimg 
import numpy
import PIL
import base64, io, IPython
from PIL import Image as PILImage
%matplotlib inline
```

```python
#importing and viewing all images from folder
img_filenames = sorted(os.listdir('The-Data-Visualization-Workshop\Datasets\images')) 
imgs = [mpimg.imread(os.path.join('The-Data-Visualization-Workshop\Datasets\images', img_filename)) for img_filename in img_filenames]
fig, axes = plt.subplots(1, 4) 
fig.figsize = (6, 6) 
fig.dpi = 150 
axes = axes.ravel() 
# Specify labels 
labels = ['coast', 'beach', 'building', 'city at night'] 
# Plot images 
for i in range(len(imgs)): 
    axes[i].imshow(imgs[i]) 
    axes[i].set_xticks([]) 
    axes[i].set_yticks([]) 
    axes[i].set_xlabel(labels[i], color='white')
```

```python
#Converting image to base64 so image file no longer needs to exist
#If sending your code to someone who can't (or doesn't want to) accept images
#Here, we covert to base64, then visualize it from that code (not from the file)
image = PILImage.open("The-Data-Visualization-Workshop\Datasets\images\photo-1.jpg")
output = io.BytesIO()
image.save(output, format='PNG')
encoded_string = base64.b64encode(output.getvalue()).decode()
html = '<img src="data:image/png;base64,{}"/>'.format(encoded_string)
IPython.display.HTML(html)
```
